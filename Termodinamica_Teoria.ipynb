 <a name="Calor"></a>
# Calor:

Cuando un sistema cambia de temperatura debido a transferencia de calor $Q$, tenemos

- Si el sistema **absorbe calor** $\to Q>0, \quad$   se calienta $\quad\to \Delta T>0$
- Si el sistema **cede calor** $\ \ \ \to Q<0, \quad\ $   se enfria $\quad\ \to \Delta T<0$

$$\boxed{Q=m\ c\ \Delta T}$$

donde: 

- c: calor específio: 

---
---

Si hay una transición de fase, entonces el calor absorbido (o cedido) es usado para romper (crear) las ligaciones atómicas, pero no para aumentar (disminuir) la temperatura.

$$Q=m\ L$$

- $L$: calor latente

 <a name="Estado termodinámico"></a>
# Estado termodinámico
  
Todo sistema **macroscópico** está descrito termodinámicamente por cierto conjunto de varíables, llamadas de variables termodinámicas o **variables de estado**.


<img align="right" src="https://ka-perseus-images.s3.amazonaws.com/da0ccdea38454a7a277614eecdbea00a543f8537.png" width="300" >

Para fluidos (gases o líquidos), las variables más comunes son: 
- $P$: presión

- $T$: temperatura

- $V$: volumen

hay muchas otras variables de estado, como por ejemplo:

- $U$: energía interna del sisterna

- $\rho$: densidad

- $S$: entropía (ver slides abajo)

- ...


No todas las variables termodinámicas son independientes.

Para muchos fluidos, apenas dos variables son independientes y las demás pueden escribirse como una función de ellas.

Ejemplo: 
  - si escogemos $P$ y $V$ como variables independientes, entonces $T$ podrá ser obtenido como una función de ellas.
  - Ese es el caso de un gás ideal, en el que $T$ es proporcional al producto de $P$ con $V$:
  $$P\ V= n\ R\ T$$


Se dice que el sistema está en un **estado termodinámico** cuando sus variables termodinámicas estan fijas. ver figura!

 <a name="Procesos termodinámicos"></a>
# Procesos termodinámicos


Se dice que el sistema pasa por un **proceso termodinámico** (o sufre una **transformación termodinámica**)  cuando al menos una de sus variables termodinámicas cambia.


Los procesos más conocidos son:

- isobárico: presión constante

- isocórico: volumen constante

- isotérmico: temperatura constante

- adiabático: sin transferencia de calor

 <a name="Ley cero de la termodinámica"></a>
# Ley cero de la termodinámica

[comment]: <> (<img src="https://solar-energia.net/uploads/cms/79/1/ley-cero-termodinamica.jpg" width="300">)



### Definiciones previas:

- Una **pared diatérmica** es aquella que **permite** el flujo de calor entre dos sistemas.
  - Se dice que los sitemas están en **contacto térmico** 
  - Ejemplo de pared diatérmica: paredes metálicas

- Una **pared adiabática** es aquella que **no permite** el flujo de calor entre dos sistemas.
  - Se dice que los sitemas están **aislados térmicamente** 
  - Ejemplo de pared adiabática: paredes de icopor, madera.



- Dos sistemas están en ** equilibrio térmico**, si al colocarlos en contacto térmico (através de una pared diatérmica) no hay flujo de calor entre ellos. Se dice que los sistemas están a la **misma temperatura**.

- Si dos sistemas no están en equilibrio térmico (están a temperaturas diferentes), entonces al colocarlos en contacto térmico habrá flujo de calor entre ellos hasta que eventualmente lleguen al equilíbrio térmico.

<img align="right" src="https://fdelarge.files.wordpress.com/2014/10/250dc-equilibrio.gif?w=458&h=344" width="300">

### Ley cero de la termodinámica

Si dos sistemas están por separado en equilibrio con un tercero, entonces también deben
estar en equilibrio entre ellos.

Esta ley da el principio básico de funcionamiento de un termómetro:

- se define una escala arbitraria (Celcius, Fahrenheit, Kelvin, etc)  através de **sistemas padrones** (sistema C) como el: hielo a $0^oC$, agua hirviendo a $100^oC$.

- se **calibra el termometro** (sistema B) al colocarlo en contacto térmico con los sistemas padrones (C).

- se mide la temperatura del sistema deseado (sistema A), al colocarlo en contact térmico con el termómetro.

# 1ra de la termodinámica:

Trata sobre la conservación de la energía: dice que la energía de un sistema sólo puede cambiar si sobre él es realizado trabajo o hay transferencia de calor.

$$\boxed{\Delta U = Q - W}$$

OBS: la convención sobre trabajo positivo o negativo está en la imagen. Algunos libros usan la convención opuesta, en cuyo caso la primeria ley seria $\Delta U = Q + W$

- **positivo** si el volumen está **aumentando**,

 <a name="Máquina térmica"></a>
# Máquina térmica

Dispositivo que convierte energía térmica en otras formas útiles de energía, como la energía eléctrica y/o mecánica.

Las máquinas térmicas funcionan en procesos cíclicos, durante los cuales:

- se absorbe calor de una fuente a alta temperatura,

- la máquina realiza un trabajo,

- se libera calor a una fuente a temperatura más baja.

La **eficiéncia** de una máquina térmica es:

$$\eta = \frac{W}{Q_\rm{absorbido}} =\frac{W}{Q_1}$$

 <a name="Refrigerador"></a>
# Refrigerador

Dispositivo cuyo objetivo es extraer calor de un cuerpo a una cierta temperatura y cederlo a otro que se encuentra a una temperatura superior.

<img align="right" src="http://www2.montes.upm.es/dptos/digfa/cfisica/termo2p/refrig_files/refrig.gif " width="400" >

- Absorbe una cantidad de calor Q2 de un foco frío a una temperatura T2

- Consume una cantidad de trabajo W

- Cede una cantidad de calor Q1 a un foco caliente a una temperatura T1

La **eficiéncia** de un refrigerador es:

$$\eta = \frac{Q_\rm{absorbido}}{|W|} =\frac{Q_2}{|W|}$$

# 2da de la termodinámica, procesos reversibles e irreversibles

### Enunciado de Planck-Kelvin (máquina térmica)

Es imposible que una máquina (térmica) *convierta todo el calor que absorbe en trabajo* durante un proceso cíclico.


### Enunciado de Clausius (refrigeradores)
No es posible el paso de calor de un cuerpo frío a uno caliente sin el consumo de trabajo.

 <a name="Entropía"></a>
# Entropía

$$dS= dQ/T$$

- En un proceso **irreversible**, la **entropía** total del Universo (sistema aislado) siempre **aumenta**. 

- En un proceso **reversible**, la **entropía** total de un Universo permanece **constante**.

Luego, para todo proceso físico:
 $$ \Delta S \ge 0$$

- Se dice que la entropía determina la dirección de la flecha del tiempo
- La entropia de un sistema ordenado es menor que la de un sistema desordenado: [La Entropía como Flecha del Tiempo - WEB]( http://hyperphysics.phy-astr.gsu.edu/hbasees/Therm/entrop.html)
